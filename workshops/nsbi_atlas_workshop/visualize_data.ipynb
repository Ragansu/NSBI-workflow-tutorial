```python
import pandas as pd
import numpy as np
import mplhep as hep
import matplotlib.pyplot as plt
hep.set_style("ATLAS")
```

```python
BASE_PATH = "./data_nsbi_tutorial/"
```

```python
nodes = [0, 5, 10]
```

```python
signals = {
    f"signal_{n:g}": pd.read_parquet(f"{BASE_PATH}/signal_{n:g}.parquet") for n in nodes
}
background = pd.read_parquet(f"{BASE_PATH}/background.parquet")
```

```python
x1_sig_0 = np.array(signals["signal_0"][f"x1"])
h_sig, bins = np.histogram(x1_sig_0, bins=50)
hep.histplot(h_sig, bins, label=f"signal_0", linewidth=2.0)

x1_sig_5 = np.array(signals["signal_5"][f"x1"])
h_sig_5, bins = np.histogram(x1_sig_5, bins=50)
hep.histplot(h_sig_5, bins, label=f"signal_5", linewidth=2.0)

x1_sig_10 = np.array(signals["signal_10"][f"x1"])
h_sig_10, bins = np.histogram(x1_sig_10, bins=50)
hep.histplot(h_sig_10, bins, label=f"signal_10", linewidth=2.0)

x1_bkg = np.array(background[f"x1"])
h_bkg, bins = np.histogram(x1_bkg, bins=50)
hep.histplot(h_bkg, bins, label=f"background", linewidth=2.0, color='green')

plt.legend()
plt.xlabel("x1", size=16)
plt.ylabel("Events", size=16)
# plt.yscale("log")
```
